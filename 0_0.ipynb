```python
# Read and save ascii files
# single spectre 
ZTF19aaksxgp = ascii.read("ZTF19aaksxgp_20190314_P60_v1.ascii")  
ZTF19aadgimr = ascii.read("ZTF19aadgimr_20190204_NOT_v1.ascii")
ZTF18abgrlpv = ascii.read("ZTF18abgrlpv_20180717_P200_v1.ascii")
# multiple spectres
ZTF19aaadwfi = []
for filename in glob.glob('ZTF19aaadwfi/*.ascii'):
    ZTF19aaadwfi.append(ascii.read(filename))
print("There are",str(np.size(ZTF19aaadwfi)),"spectres for ZTF19aaadwfi")

ZTF18aavskep = []
for filename in glob.glob('ZTF18aavskep/*.ascii'):
    ZTF18aavskep.append(ascii.read(filename))
print("There are",str(np.size(ZTF18aavskep)),"spectres for ZTF18aavskep")

ZTF18abltfho = []
for filename in glob.glob('ZTF18abltfho/*.ascii'):
    ZTF18abltfho.append(ascii.read(filename))
print("There are",str(np.size(ZTF18abltfho)),"spectres for ZTF18abltfho")

ZTF18abxbhov = []
for filename in glob.glob('ZTF18abxbhov/*.ascii'):
    ZTF18abxbhov.append(ascii.read(filename))
print("There are",str(np.size(ZTF18abxbhov)),"spectres for ZTF18abxbhov")

ZTF18acwzyor = []
for filename in glob.glob('ZTF18acwzyor/*.ascii'):
    ZTF18acwzyor.append(ascii.read(filename))
print("There are",str(np.size(ZTF18acwzyor)),"spectres for ZTF18acwzyor")
```
