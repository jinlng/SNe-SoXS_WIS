```python
import numpy as np
from scipy import signal
from itertools import product
import matplotlib.pyplot as plt
import scipy.integrate as integrate
```

```python
# Define useful functions in math
# Gaussian
def fct_Gauss(x,esp,var):
    return ((1./np.sqrt(var*2*np.pi))*np.exp(-0.5*(x-esp)**2/var))
# Delta
def delta(n):
    if n == 0:
        return 1
    else:
         return 0
# Convolution
def convolution(a,b):
    conv = []
    for i in range(len(a)):
        somme = 0
        for j in range(len(b)):
            somme += a[i-j]*b[j]
        conv.append(somme)
    return conv
```

```python
fi = []
n = 100
n_pic = 10

amp = np.ones(n_pic,dtype=int)
dis = np.arange(0,n,n_pic)

for i,j in product(range(n),range(n_pic)):
    fi.append(amp[j]*delta(i-dis[j]))

print('Amplitude:', amp)
print('Distance:', dis)

plt.figure()
plt.plot(fi)
plt.grid(True)
```

```python
sigma = 10
esperance = len(fi)/2
x = np.arange(0,len(fi),1)
fo = convolution(fi, fct_Gauss(x,esperance,sigma))
plt.figure()
plt.plot(fo)
plt.grid(True)
plt.show()
```
