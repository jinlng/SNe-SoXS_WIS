```python
import time
import numpy as np
from scipy import signal
from itertools import product
import matplotlib.pyplot as plt
import scipy.integrate as integrate
```

```python
n = 10000
sig_in = 0.01
cste = np.sqrt(-np.log(0.37703)*8)
wavelength = np.linspace(3000,8000,n)

R_cste = 4500
dLam_cste = wavelength / R_cste
sig_g_cste = dLam_cste / np.sqrt(8*np.log(2))
sig_out_cste = np.sqrt(sig_in**2 + sig_g_cste**2)
d_cste = cste * sig_out_cste

R_var = np.linspace(4000,5000,n) 
lam = np.linspace(3000,8000,n)
dLam_var = lam/R_var
sig_g_var = dLam_var / np.sqrt(8*np.log(2))
sig_out_var = np.sqrt(sig_in**2 + sig_g_var**2)
d_var = cste * sig_out_var

plt.figure(figsize=[16,9])
plt.suptitle('Limit resolution distance en fonction de wavelength',fontsize=20)
plt.plot(wavelength,d_cste,'--',label='Constant resolution R=4500')
plt.plot(wavelength,d_var,label='Linear increasing resolution R=[4000,5000]')
plt.xlabel('wavelength ($\AA$)',fontsize=20)
plt.ylabel('Distance for limit resolution ($\AA$)',fontsize=20)
plt.legend()
plt.grid(True)
plt.show()
```
