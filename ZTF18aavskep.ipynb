```python
import os
import glob
import pylab
import numpy as np
from astropy.io import ascii
import matplotlib.pyplot as plt

ZTF18aavskep = []
for filename in glob.glob('ZTF18aavskep/*.ascii'):
    ZTF18aavskep.append(ascii.read(filename))
print("There are",str(np.size(ZTF18aavskep)),"spectres for ZTF18aavskep")

n = np.size(ZTF18aavskep)
for i in range(n):
    pylab.plot(ZTF18aavskep[i][0][:],ZTF18aavskep[i][1][:])
    pylab.show()
```
